```python
import pandas as pd
from copulas.univariate import GaussianKDE
from copulas.multivariate import GaussianMultivariate, VineCopula
```

```python
def read_dataset(name='data/factor_capacidad.csv'):
    df = pd.read_csv(name,index_col=0,parse_dates=True)
    return df
df = read_dataset()
```

```python
dist = GaussianMultivariate(distribution={
    'solar_pv':GaussianKDE,
    'solar_th':GaussianKDE,
    'wind':GaussianKDE,
})
dist.fit(df.sample(7500))
params = dist.to_dict()
print(params['correlation'])
print([a['type'] for a in params['univariates']])
```

```python
center = VineCopula('center')
regular = VineCopula('regular')
direct = VineCopula('direct')

center.fit(df.sample(5000))
regular.fit(df.sample(5000))
direct.fit(df.sample(5000))

params_c = center.to_dict()
params_c.keys()
```
