```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [2017,2018,2019,2020,2021,2022]
y = [736.5,6293.81,8411.56,8874.31,6304.37,11418.91]

plt.bar(x,y)
plt.ylabel('Investment in clean energy (million USD)')
plt.grid(True, axis='y')
plt.ylim(bottom=0)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
```

```python
labels = ['Hydro', 'Wind', 'Solar PV', 'Solar Thermal', 'Renewable waste', 'Other']
sizes = [25273,62569,37332,4694,846,3590+17]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']

# Create the pie chart with improvements
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(sizes, colors=colors, autopct='%1.1f%%',
                                  startangle=140, wedgeprops=dict(width=0.4), pctdistance=0.85)



# Improve the layout
plt.setp(autotexts, size=7, weight="bold", color="white")
plt.setp(texts, size=10)
ax.legend(wedges, labels, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')

# Display the improved pie chart
plt.show()


# plt.pie(y)
# plt.legend(x,loc='best')
# # plt.ylabel('Investment in clean energy (million USD)')
# # plt.grid(True, axis='y')
# # plt.ylim(bottom=0)
# for spine in plt.gca().spines.values():
#     spine.set_visible(False)
```
